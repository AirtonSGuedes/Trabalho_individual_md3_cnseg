# Relatório de Análise Semanal do Projeto de Desenvolvimento de Software

## Introdução
Este relatório tem como objetivo fornecer uma análise abrangente do progresso do projeto de desenvolvimento de software ao longo de uma semana. Os dados apresentados são fundamentais para identificar padrões, tendências e áreas de melhoria, contribuindo assim para o sucesso contínuo do projeto.



## Criando um DataBase

```
import pandas as pd
import numpy as np


dados = {
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'sábado', 'Domingo'],
    'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2]
}
```

## Gerando o relatorio


```
# DataFrame
db = pd.DataFrame(dados)

# Cálculos
Horas_tot = db['Horas Trabalhadas'].sum()
Bugs_corrigidos = db['Bugs Corrigidos'].sum()
Tarefas_tot_concluidas = db['Tarefas Concluídas'].sum()
horas_media = db['Horas Trabalhadas'].mean()
medias_bugs = db['Bugs Corrigidos'].mean()
media_tarefas =db['Tarefas Concluídas'].mean()

```

##  criando o grafico pra exibir os resultados






```
import matplotlib.pyplot as plt
resultados = pd.DataFrame({
    'Métrica': ['Horas Totais', 'Bugs Corrigidos', 'Tarefas Totais', 'Média de Horas', 'Média de Bugs', 'Média de Tarefas'],
    'Valor': [Horas_tot, Bugs_corrigidos, Tarefas_tot_concluidas, horas_media, medias_bugs, media_tarefas]
})
# Gráficos de Barras
plt.figure(figsize=(10,6))
plt.bar(resultados['Métrica'], resultados['Valor'], color='skyblue')
plt.title('Análise de Métricas')
plt.xlabel('Métrica')
plt.ylabel('Valor')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

display(resultados)
```

## Vamos analisar as medias de bugs e de tarefas e obsever qual foi o dia menos produtivo



```
plt.figure(figsize=(10,6))
plt.scatter(db['Dia'], db['Bugs Corrigidos'], label='Média de Bugs', color='red')
plt.scatter(db['Dia'], db['Tarefas Concluídas'], label='Média de Tarefas', color='blue')
plt.xlabel('Dias da Semana')
plt.ylabel('Valor')
plt.title('Gráfico de Dispersão: Média de Bugs e Tarefas por Dia')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Agora Vamos Analisar a horas totais e tarefas concluidas

```

plt.figure(figsize=(10,6))
plt.bar(db['Dia'], db['Horas Trabalhadas'], label='Horas Trabalhadas', color='skyblue')
plt.bar(db['Dia'], db['Tarefas Concluídas'], label='Tarefas Concluídas', color='red')
plt.title('Gráfico de Colunas: Horas Trabalhadas e Tarefas Concluídas por Dia')
plt.xlabel('Dias da Semana')
plt.ylabel('Valor')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
